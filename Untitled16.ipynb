```python
import pandas as pd
```

```python
df=pd.read_csv("carprices.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
df
```

```python

X = [df['Age(yrs)']]
y= [df['Sell Price($)']]
```

```python
plt.scatter(X,y,data="df")
```

```python
dummies=pd.get_dummies(df['Car Model'])
dummies
```

```python
df_with_dummies=pd.concat([df,dummies],axis="columns")
```

```python
df_with_dummies
```

```python
df_with_dummies.drop(['Car Model','Audi A5'],axis='columns',inplace=True)
```

```python
y=df_with_dummies['Sell Price($)']
```

```python
X = df_with_dummies.drop(['Sell Price($)','Mileage'],axis='columns')
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[4,1,0]])
```
